# Betweenness Centrality
Betweenness centrality is a way of detecting the amount of influence a node has over the flow of information in a graph. It is often used to find nodes that serve as a bridge from one part of a graph to another. The algorithm calculates shortest paths between all pairs of nodes in a graph.

```python
import networkx as nx
import triplea.service.graph.export.export as gexport
import triplea.service.graph.analysis.ganalysis as ganaliz
```

```python
G = nx.read_graphml(r"C:\Users\Bitaraf\Desktop\new\triples.graphml")
```

```python
dcs = ganaliz.sorted_betweenness_centrality(G)
```

you can save the pandas Series of betweenness centrality values (dcs) to a file in Python:

```python
dcs.to_csv('betweenness_centrality.csv')
```

The key steps for draw scatter plot:

- Extract the betweenness centrality series into a list for the x values
- Generate a range of nodes for the y values
- Plot the scatter plot
- Invert the y-axis so the nodes with highest betweenness are at the top

This will generate the scatter plot with the nodes ordered vertically by their betweenness centrality value.

```python
import matplotlib.pyplot as plt

# Extract betweenness centrality values into lists
x = list(dcs.values) 
# y = range(len(dcs))
y = list(range(len(dcs)))[::-1] # reverse order 

plt.scatter(x, y)
plt.xlabel('Betweenness Centrality')  
plt.ylabel('Node')
plt.title('Betweenness Centrality Scatter Plot')

# Invert y-axis so highest Betweenness nodes are at top
# plt.gca().invert_yaxis()

plt.show()
```
